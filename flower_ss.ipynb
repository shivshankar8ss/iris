```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species')
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
sns.pairplot(df, hue='Species')
plt.show()
```

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
df_melted = df.melt(id_vars='Species', value_vars=features, var_name='Feature', value_name='Value')

plt.figure(figsize=(12, 8))
sns.histplot(data=df_melted, x='Value', hue='Species', multiple='stack', bins=20)
plt.title('Histogram of Features Across Species')
plt.show()
```

```python
# Boxplot for each feature across the species
plt.figure(figsize=(12, 8))
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Boxplot of Sepal Length Across Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Boxplot of Sepal Width Across Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Boxplot of Petal Length Across Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Boxplot of Petal Width Across Species')
plt.show()
```

```python
# Violinplot for each feature across species
plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Violin Plot of Sepal Length Across Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Violin Plot of Sepal Width Across Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Violin Plot of Petal Length Across Species')
plt.show()

plt.figure(figsize=(12, 8))
sns.violinplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Violin Plot of Petal Width Across Species')
plt.show()
```
